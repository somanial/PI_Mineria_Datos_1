```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargo el dataset que limpie en el paso anterior
df_clean = pd.read_json("../data/processed/streaming_users_clean.json")

# Configuro un estilo para los gráficos
sns.set_theme(style="whitegrid")
print("¡Librerías cargadas y datos listos para graficar!")
```

```python
# Creación de una figura con dos espacios para gráficos univariados
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Distribución de Edades (Histograma)
sns.histplot(df_clean['age'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribución de Edades de los Usuarios')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Frecuencia')

# Gráfico 2: Cantidad de Usuarios por Plan de Suscripción (Gráfico de Barras)
sns.countplot(x='subscription_plan', data=df_clean, ax=axes[1], palette='pastel', order=df_clean['subscription_plan'].value_counts().index)
axes[1].set_title('Cantidad de Usuarios por Plan de Suscripción')
axes[1].set_xlabel('Plan de Suscripción')
axes[1].set_ylabel('Cantidad de Usuarios')

plt.tight_layout()
plt.show()
```

### Interpretación del Análisis Univariado

1. **Distribución de Edades**: El histograma revela cómo se distribuye la base de usuarios según su edad. Esto permite identificar si la plataforma concentra un público joven, adulto o si mantiene una distribución uniforme, lo cual es clave para definir el perfil sociodemográfico de la muestra y orientar las estrategias de contenido.
2. **Planes de Suscripción**: El gráfico de barras permite contrastar la popularidad de los distintos niveles de suscripción (Basic, Standard, Premium). Conocer el volumen de usuarios en cada categoría ayuda a entender el comportamiento de consumo inicial y la distribución del modelo de ingresos de la plataforma antes de cruzarlo con otras métricas.

```python
# Creación de una figura con dos espacios para gráficos bivariados
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 3: Edad vs Tiempo de Visualización Mensual (Scatter Plot / Dispersión)
sns.scatterplot(x='age', y='monthly_watch_time_mins', data=df_clean, ax=axes[0], alpha=0.6, color='purple')
axes[0].set_title('Relación entre Edad y Tiempo de Visualización')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Tiempo de Visualización Mensual (min)')

# Gráfico 4: Plan de Suscripción vs Tickets de Soporte (Boxplot / Diagrama de Caja)
sns.boxplot(x='subscription_plan', y='customer_support_tickets', data=df_clean, ax=axes[1], palette='Set2')
axes[1].set_title('Tickets de Soporte Técnico según el Plan')
axes[1].set_xlabel('Plan de Suscripción')
axes[1].set_ylabel('Cantidad de Tickets de Soporte')

plt.tight_layout()
plt.show()
```

### Interpretación del Análisis Bivariado

1. **Edad vs. Tiempo de Visualización**: Al cruzar estas dos variables cuantitativas mediante un gráfico de dispersión, buscamos identificar patrones de consumo o tendencias generacionales. Esto permite analizar si el nivel de enganche (engagement) con el contenido varía de forma lineal o dispersa según los diferentes rangos etarios de la muestra.
2. **Plan de Suscripción vs. Tickets de Soporte**: El diagrama de caja (boxplot) nos permite evaluar el comportamiento de la demanda de asistencia técnica según el tipo de cliente. Es fundamental para determinar si los usuarios de planes Premium (que aportan un mayor ingreso) experimentan más fricciones o si, por el contrario, la tasa de incidencias se mantiene uniforme entre todos los niveles de suscripción.

```python
# Gráfico 5: Género Favorito vs Tiempo de Visualización segmentado por Plan de Suscripción (Multivariado)
plt.figure(figsize=(12, 7))
sns.barplot(x='favorite_genre', y='monthly_watch_time_mins', hue='subscription_plan', data=df_clean, palette='Set1', ci=None)

plt.title('Tiempo de Visualización Mensual por Género Favorito y Plan de Suscripción')
plt.xlabel('Género Favorito')
plt.ylabel('Tiempo de Visualización Mensual Promedio (min)')
plt.xticks(rotation=45)
plt.legend(title='Plan de Suscripción', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Interpretación del Análisis Multivariado

1. **Género, Tiempo de Visualización y Plan de Suscripción**: Este análisis tridimensional permite evaluar simultáneamente las preferencias de contenido, la intensidad del consumo y el nivel comercial del usuario. Cruzar estas tres variables ayuda a descubrir si ciertos géneros específicos retienen más a los usuarios de mayor valor (Premium), lo cual aporta evidencia estratégica crucial para la toma de decisiones sobre la adquisición de licencias de contenido y la personalización de la plataforma.
