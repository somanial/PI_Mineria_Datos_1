Este cuaderno consolida los hallazgos críticos obtenidos a lo largo de las etapas de inspección, limpieza, análisis exploratorio (EDA) y procesamiento de variables mediante PCA. El objetivo es estructurar las conclusiones alineadas a las preguntas de negocio, diferenciando rigurosamente la evidencia, la interpretación y la conclusión final.

## Bloque 1: Perfil Sociodemográfico y Consumo Estratégico (Planes Premium y Retención)

*   **Evidencia (El Dato):** El gráfico bivariado y el multivariado demostraron que los usuarios del Plan Premium no poseen una concentración exclusiva en un rango etario alto, sino que están distribuidos, pero exhiben los mayores niveles de `monthly_watch_time_mins` (minutos de visualización mensuales) y, en paralelo, la mayor tasa de interacción con el soporte técnico (`customer_support_tickets`).
*   **Interpretación (El Significado):** Los clientes Premium son usuarios de "alto valor" y "alta fricción". Esto significa que consumen intensivamente la plataforma, lo que los vuelve más propensos a experimentar problemas técnicos o a demandar atención. Su retención no depende de su edad, sino de la velocidad y eficiencia con la que resolvemos sus reclamos de soporte.
*   **Conclusión (La Acción):** Se concluye que para mitigar el Churn (abandono) en los segmentos que más ingresos generan, la empresa debe priorizar una cola de atención rápida o "Fast Track" en el sistema de tickets para usuarios Premium. El tiempo de visualización es un indicador de lealtad que debe protegerse mediante soporte proactivo.

## Bloque 2: Consistencia del Dataset y Calidad de Datos

*   **Evidencia (El Dato):** Al graficar el consumo por género y plan en el Notebook 03, se detectaron barras duplicadas y categorías fragmentadas debido a problemas de inconsistencia en el formateo de texto (por ejemplo, diferencias de mayúsculas/minúsculas y caracteres).
*   **Interpretación (El Significado):** El análisis exploratorio funcionó como un mecanismo de auditoría interna. Evidenció que una limpieza genérica automatizada es insuficiente cuando existen fallas de carga manual en las variables categóricas de origen.
*   **Conclusión (La Acción):** Se concluye y valida la decisión metodológica de haber retornado al Notebook 02 para aplicar un mapeo explícito (`.map()`). Esto garantiza que los futuros modelos de Machine Learning no interpreten "Premium" y "premium" como dos entidades de negocio distintas, asegurando la robustez e integridad de las predicciones.

## Bloque 3: Estructura de Dimensiones y Viabilidad de PCA

*   **Evidencia (El Dato):** El Análisis de Componentes Principales (PCA) en el Notebook 04 arrojó que la varianza explicada se distribuye de manera casi equitativa entre los tres componentes (aproximadamente 33.5% para PC1, 33.3% para PC2 y 33.1% para PC3).
*   **Interpretación (El Significado):** Ninguna de las variables numéricas (`age`, `monthly_watch_time_mins`, `customer_support_tickets`) domina matemáticamente la estructura de los datos ni está fuertemente correlacionada de forma lineal con las otras. Cada variable aporta una dimensión de información completamente independiente y única.
*   **Conclusión (La Acción):** Desde el punto de vista metodológico, se concluye que **no es recomendable comprimir drásticamente el dataset eliminando componentes** si se busca entrenar un modelo que requiera alta precisión, ya que prescindir de un componente implicaría perder un tercio completo de la variabilidad de la información. Para la etapa de modelado e informe en Streamlit, se sugiere mantener las variables originales estandarizadas o retener la totalidad de los componentes para salvaguardar el poder predictivo.
